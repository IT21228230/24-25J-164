```python
import os
import numpy as np
import tensorflow as tf
import shutil
from tensorflow.keras import layers, models
from tensorflow.keras.applications import VGG16
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Model
import pathlib
```

```python
original_data_dir = 'D:/Source'  
base_dir = 'D:/Base'
train_dir = os.path.join(base_dir, 'train')
val_dir = os.path.join(base_dir, 'validation')
test_dir = os.path.join(base_dir, 'test')
```

```python
# Create new directories for train, validation, and test splits
for split_dir in [train_dir, val_dir, test_dir]:
    if os.path.exists(split_dir):
        shutil.rmtree(split_dir)  # Clear existing split data if rerunning the code
    os.makedirs(split_dir)

# Get all classes from the dataset
class_names = [item.name for item in pathlib.Path(original_data_dir).glob('*') if item.is_dir()]

# Split each class into train, validation, and test sets
for class_name in class_names:
    class_path = os.path.join(original_data_dir, class_name)
    images = os.listdir(class_path)
    images = [os.path.join(class_path, img) for img in images if img.lower().endswith(('.png', '.jpg', '.jpeg'))]
    
    if len(images) == 0:
        print(f"No images found in {class_path}, skipping this class.")
        continue

    print(f"Found {len(images)} images in class '{class_name}'.")

    train_images, test_images = train_test_split(images, test_size=0.2, random_state=42)
    train_images, val_images = train_test_split(train_images, test_size=0.25, random_state=42)  # 0.25 * 0.8 = 0.2

    # Define class directories in train, val, and test
    os.makedirs(os.path.join(train_dir, class_name), exist_ok=True)
    os.makedirs(os.path.join(val_dir, class_name), exist_ok=True)
    os.makedirs(os.path.join(test_dir, class_name), exist_ok=True)

    # Move images to respective directories
    for img in train_images:
        shutil.copy(img, os.path.join(train_dir, class_name))
    for img in val_images:
        shutil.copy(img, os.path.join(val_dir, class_name))
    for img in test_images:
        shutil.copy(img, os.path.join(test_dir, class_name))

print("Dataset splitting completed.")
```

```python
img_size = (224, 224)  # VGG16 input size
batch_size = 32
```

```python
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)

val_test_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical'
)

val_generator = val_test_datagen.flow_from_directory(
    val_dir,
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical'
)

test_generator = val_test_datagen.flow_from_directory(
    test_dir,
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical'
)
```

```python
base_model = VGG16(include_top=False, input_shape=(224, 224, 3), weights='imagenet')
for layer in base_model.layers:
    layer.trainable = False
```

```python
x = base_model.output
x = Flatten()(x)
x = Dense(512, activation='relu')(x)
x = Dense(len(class_names), activation='softmax')(x)
```

```python
model = Model(inputs=base_model.input, outputs=x)

model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.0001),
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
history = model.fit(
    train_generator,
    validation_data=val_generator,
    epochs=20,
    steps_per_epoch=train_generator.samples // train_generator.batch_size,
    validation_steps=val_generator.samples // val_generator.batch_size
)
```

```python
test_loss, test_accuracy = model.evaluate(test_generator)
print(f"Test accuracy: {test_accuracy:.4f}")
```

```python
from sklearn.metrics import roc_curve, auc, roc_auc_score
from sklearn.preprocessing import label_binarize
from sklearn.metrics import RocCurveDisplay
from tensorflow.keras.models import load_model

model = load_model('D:/Disease Classification/trained_model.h5')
y_pred_proba = model.predict(test_generator)  # shape: (n_samples, n_classes)

# Binarize the labels for multi-class ROC AUC
n_classes = y_pred_proba.shape[1]
y_true = label_binarize(test_generator.classes, classes=range(n_classes))
fpr = dict()  # False Positive Rate
tpr = dict()  # True Positive Rate
roc_auc = dict()  # Area Under the Curve

for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(y_true[:, i], y_pred_proba[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])

plt.figure(figsize=(10, 8))
for i in range(n_classes):
    plt.plot(fpr[i], tpr[i], label=f"Class {i} (AUC = {roc_auc[i]:.2f})")

# Plotting diagonal line (random classifier)
plt.plot([0, 1], [0, 1], 'k--', label='Random')

plt.title('Multi-Class ROC Curve')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend(loc='best')
plt.grid()
plt.show()

```

```python
# Compute macro-average ROC AUC
all_fpr = np.unique(np.concatenate([fpr[i] for i in range(n_classes)]))
mean_tpr = np.zeros_like(all_fpr)

for i in range(n_classes):
    mean_tpr += np.interp(all_fpr, fpr[i], tpr[i])

mean_tpr /= n_classes
macro_auc = auc(all_fpr, mean_tpr)

# Compute micro-average ROC AUC
micro_auc = roc_auc_score(y_true, y_pred_proba, average="micro")

print(f"Macro-average AUC: {macro_auc:.2f}")
print(f"Micro-average AUC: {micro_auc:.2f}")
```
